<h3>Download trained pipeline</h3>

To download trained pipeline use a command such as,

python -m spacy download en_core_web_sm

This downloads the small (sm) pipeline for english language

Further instructions on : https://spacy.io/usage/models#quickstart

```python
nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
nlp.pipeline
```

sm in en_core_web_sm means small. There are other models available as well such as medium, large etc. Check this: https://spacy.io/usage/models#quickstart

```python
doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token, " | ", spacy.explain(token.pos_), " | ", token.lemma_)

# token.pos_ تعني Part of Speech (نوع الكلمة)، وهي تعيد رمزاً مثل VERB أو PROPN.
#spacy.explain() هي دالة سحرية تحول هذا الرمز إلى شرح مفهوم (مثلاً تحول VERB إلى verb).
# token.lemma_: تعني Lemmatization (الرد إلى الأصل). وهي تعيد الكلمة إلى شكلها الأساسي في القاموس (مثلاً: "ate" تصبح "eat").
```

<h3>Named Entity Recognition</h3>
"التعرف على الكيانات المسماة" (Named Entity Recognition - NER).

هنا، البرنامج لا يكتفي بفهم نوع الكلمة (اسم أو فعل)، بل يفهم "ماذا تمثل هذه الكلمة في العالم الحقيقي".

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy

displacy.render(doc, style="ent")

#displacy: هي أداة عرض مدمجة في spaCy تجعل النتائج تظهر بشكل بصري رائع (تلوين الكلمات المكتشفة).
#style="ent": نخبر الأداة أننا نريد تلوين "الكيانات" (Entities).
```

```python
print(spacy.explain("ORG")) #إذا أردت أن تعرف ماذا يعني الرمز ORG
```

NER: قدرة الحاسوب على تحديد الأسماء، الشركات، الدول، والمبالغ المالية.

ORG: رمز للشركات والمؤسسات.

MONEY: رمز للعملات والأرقام المالية.

displacy: أفضل طريقة لعرض نتائج تحليل النصوص للمستخدمين أو في التقارير.

<h3>Trained processing pipeline in French</h3>

```python
nlp = spacy.load("fr_core_news_sm") #هذا الأمر يخبر مكتبة spaCy بتحميل "المخ" أو "النموذج الإحصائي" الخاص باللغة الفرنسية
```

You need to install the processing pipeline for french language using this command,

python -m spacy download fr_core_news_sm

```python
nlp = spacy.load("fr_core_news_sm")
```

هذا الموديل يأتي محملاً "بأنابيب معالجة" جاهزة. بمجرد تحميله، يمكنك كتابة 
print(nlp.pipe_names) وستجد المكونات التالية مفعلة تلقائياً:

tok2vec: لتحويل الكلمات لمتجهات رقمية.

tagger: لتحديد نوع الكلمات (اسم، فعل، حرف).

parser: لفهم العلاقات النحوية بين الكلمات.

ner: للتعرف على الأماكن والأسماء بالفرنسية.

```python
doc = nlp("Tesla Inc va racheter Twitter pour $45 milliards de dollars")
for ent in doc.ents:
    print(ent.text, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
for token in doc:
    print(token, " | ", token.pos_, " | ", token.lemma_)
```

<h3>Adding a component to a blank pipeline</h3>

```python
source_nlp = spacy.load("en_core_web_sm")

nlp = spacy.blank("en")
nlp.add_pipe("ner", source=source_nlp)
nlp.pipe_names
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
