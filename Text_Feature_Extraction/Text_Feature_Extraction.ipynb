```python
def CheckReplies(tweet):
    # الخطوة 1: التأكد أن المدخل نص وليس قيمة فارغة (NaN)
    if not isinstance(tweet, str) or tweet.strip() == "":
        return '***NoReply***'
        
    FirstReply = ''
    if '@' in tweet:
        parts = tweet.split('@')
        # الخطوة 2: التأكد من وجود جزء ثانٍ بعد علامة @
        if len(parts) > 1:
            words_after_at = parts[1].split()
            # الخطوة 3: التأكد أن الجزء الثاني يحتوي على كلمات (ليس فراغاً)
            if len(words_after_at) > 0:
                FirstReply = words_after_at[0]
                # تنظيف الكلمة المستخرجة
                FirstReply = FirstReply.replace(':', '').replace(')', '').lower()
                return FirstReply
                
    return '***NoReply***'

def FirstLink(tweet):
    # التأكد من صحة المدخلات
    if not isinstance(tweet, str) or tweet.strip() == "":
        return '***NoLink***'
        
    target_link = 'http://t.co/'
    if target_link in tweet:
        # البحث عن الكلمات التي تحتوي على الرابط وتخزينها في قائمة
        links = [i for i in tweet.split() if target_link in i]
        # الخطوة 4: التأكد أن القائمة ليست فارغة قبل محاولة الوصول للعنصر [0]
        if len(links) > 0:
            return links[0]
            
    return '***NoLink***'
```

```python
%%time
# data['TweetLetters'] = data['text'].apply(lambda x : len(x))#يحسب عدد حروف التغريدة (طول النص)
# data['TweetWords'] = data['text'].apply(lambda x : len(x.split())) #يحسب عدد الكلمات (عن طريق تقطيع النص بالمسافات ثم عده).
# 1. حساب طول الحروف والكلمات بطريقة متجهة
data['TweetLetters'] = data['text'].str.len()
data['TweetWords'] = data['text'].str.split().str.len()
data['ReplyNumber'] = data['text'].apply(lambda n : n.count('@')) #حسب كم مرة ظهرت علامة @. (عدد الردود/الإشارات)
data['FirstReply'] = data['text'].apply(lambda T :  CheckReplies(T))
data['LinkExistence'] = data['text'].apply(lambda x: r'http://t.co/' in x )#عمود منطقي (True/False) يخبرنا: هل تحتوي التغريدة على رابط أصلاً؟
data['FirstLink'] = data['text'].apply(lambda x: FirstLink(x))


```
